# Half-moon Example

This example shows how to use `pydatapull` to generate variants of the half-moon problem.

In `pydatapull`, the half-moon classes are defined by the following equations.

The half-moons are generated by sampling from normal distributions with means and standard deviations given by
$$
\mu_{xa}(t_a),\ \sigma_{xa}(t_a),\ \mu_{ya}(t_a),\ \sigma_{ya}(t_a)
$$
for class A and 
$$
\mu_{xb}(t_b),\ \sigma_{xb}(t_b),\ \mu_{yb}(t_b),\ \sigma_{yb}(t_b)
$$
for class B.

The first class has means defined by
$$
\begin{align*}
\mu_{xa}(t_a) &= \frac{-r}{2} + r \cos(t_a) \\
\mu_{ya}(t_a) &= d + r \sin(t_a)
\end{align*}
$$
where
$$
0 < t_a < \pi .
$$

The second class has means defined by
$$
\begin{align*}
\mu_{xa}(t_a) &= \frac{r}{2} + r \cos(t_b) \\
\mu_{ya}(t_a) &= -d + r \sin(t_b).
\end{align*}
$$
where
$$
\pi < t_b < 2\pi .
$$

All standard deviations are constant,
$$
\sigma_{xa} = \sigma_{xb} = \sigma_{ya} = \sigma_{yb} = \sigma
$$

Clusters are created for specific values of $t_a$ and $t_b$ with $N_S$ clusters and $N_P$ points per cluster.

## Imports

Here we import `generate_half_moon_dataset` to generate the dataset and `scatter_plot_dataset` to visualize it.

```python
from pydatapull.basicml import generate_half_moon_dataset
from pydatapull.visualizers import scatter_plot_dataset

import matplotlib.pyplot as plt

plt.rcParams["text.usetex"] = True
plt.rcParams["font.size"] = 14
plt.rcParams["font.family"] = "serif"
plt.rcParams["mathtext.fontset"] = "dejavuserif"
```

## Generate a Classic Half Moon

Here we generate a classic half moon. We can make it smoother by increasing the clusters per class and decreasing the noise (sigma). We use `scatter_plot_dataset` to visualize it.

```python
x_features, y_labels, _ = generate_half_moon_dataset(
    n_samples_per_class=1000,
    n_clusters_per_class=100,
    radius_r=1,
    separation_offset_d=-0.2,
    sigma=0.1,
)

_ = scatter_plot_dataset(x_features, y_labels)
```

## Generate Variants on the Half-moon

We can generate interesting variants on the half-moon by changing the parameters.

### Smooth and Skinny Variant

```python
x_features, y_labels, _ = generate_half_moon_dataset(
    n_samples_per_class=1000,
    n_clusters_per_class=500,
    radius_r=2,
    separation_offset_d=0,
    sigma=0.01,
)

_ = scatter_plot_dataset(x_features, y_labels)
```

### Clusters within Half-moons

```python
x_features, y_labels, _ = generate_half_moon_dataset(
    n_samples_per_class=1000,
    n_clusters_per_class=10,
    radius_r=10,
    separation_offset_d=-2,
    sigma=0.4,
)

_ = scatter_plot_dataset(x_features, y_labels)
```
